### Legend

```python
co=[1,2,3,4,5,6,7,8]
a=[50,34,34,56,22,12,67,89]
b=[23,45,67,89,35,23,56,21]
c=[55,32,45,67,34,23,56,88]
plt.figure(figsize=(10,5))
plt.plot(co,a,color='r',label='co and a')
plt.plot(co,b,color='g',label='co and b')
plt.plot(co,c,color='b',label='co and c')
plt.legend(loc=0)
plt.show()
```

```python
co=[1,2,3,4,5,6,7,8]
a=[50,34,34,56,22,12,67,89]
b=[23,45,67,89,35,23,56,21]
c=[55,32,45,67,34,23,56,88]
plt.figure(figsize=(10,5))
plt.plot(co,a,color='g',label='co and a')
plt.plot(co,b,color='b',label='co and b')
plt.plot(co,c,color='r',label='co and c')
plt.legend(loc=0, shadow=True, fontsize='large')
```

```python
courses=['python','java','C#','Go']
revenue=[1800,1200,840,450]
```

```python
len(courses)
```

```python
a=np.arange(len(courses))
```

```python
a
```

```python
plt.bar(a,revenue,label='Revenue data',color='g')
plt.title('BAR PLOT')
plt.xticks(a,courses)
plt.ylabel('revenue')
plt.legend(loc=0)
plt.show()
```

```python
plt.barh(a,revenue,label='Revenue data ',color='b')
plt.title('BAR PLOT')
plt.yticks(a,courses)
plt.xlabel('revenue')
plt.legend(loc=0)
plt.show()
```

### Sub plot

```python
co=[1,2,3,4,5,6,7,8]
a=[50,34,34,56,22,12,67,89]
b=[23,45,67,89,35,23,56,21]
plt.subplot(2,1,1)
plt.plot(co,a,color='r',label='co and a')
plt.subplot(2,1,2)
plt.plot(co,b,color='g',label='co and b')
plt.legend(loc=0)
```

```python
a=[2014,2015,2016,2017]
b=[4000,4500,5300,4600]
c=[2800,2645,3400,3000]

plt.subplot(1,2,1)
plt.plot(a,b,color='r')
plt.subplot(1,2,2)
plt.plot(a,c,color='g')
```

```python
co=[2014,2015,2016,2017]
a=[4000,4000,5300,4600]
b=[2800,3000,2800,3400]
plt.subplot(1,2,1)
plt.plot(a,b,color='r')
plt.subplot(1,2,2)
plt.plot(a,c,color='g')
```

### Pie Chart

```python
x=[40,30,20,10,1]
y=['java','python','C#','PHP','C++']
plt.figure(figsize=(10,5))
plt.pie(x,labels=y)
```

```python
x=[40,30,20,10,1]
y=['java','python','C#','PHP','C++']
plt.figure(figsize=(10,5))
plt.pie(x,labels=y,shadow=True)
```

```python
x=[40,30,20,10,1]
y=['java','python','C#','PHP','C++']
plt.figure(figsize=(10,5))
plt.pie(x,labels=y,shadow=True,autopct='%1.1f%%')
```

```python
x=[40,30,20,10,1]
y=['java','python','C#','PHP','C++']
plt.figure(figsize=(10,5))
plt.pie(x,labels=y,shadow=True,autopct='%1.1f%%',explode=[0,0.4,0,0.2,0])
```
